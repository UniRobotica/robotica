## 1. Introdução ao Otto Ninja
O Otto Ninja é um robô bípede open-source baseado em Arduino, capaz de andar, dançar e desviar de obstáculos.

### Conceitos envolvidos:
- Robótica educacional
- Sistemas embarcados
- Controle de servomotores
- Sensores ultrassônicos

## 2. Componentes
- Arduino Nano
- 4x Servo motores SG90
- Sensor ultrassônico HC-SR04
- Buzzer
- Módulo Bluetooth (opcional)
- Bateria
- Estrutura impressa em 3D

```python
# Simulação simples de controle de servo (conceito)
import numpy as np
import matplotlib.pyplot as plt

tempo = np.linspace(0, 2, 100)
angulo = 45 * np.sin(2 * np.pi * tempo)

plt.plot(tempo, angulo)
plt.xlabel('Tempo')
plt.ylabel('Ângulo do servo')
plt.title('Movimento Oscilatório do Servo')
plt.show()
```

## 3. Montagem
1. Fixar os servos na estrutura
2. Conectar os servos ao Arduino
3. Instalar sensor ultrassônico na cabeça
4. Conectar alimentação

### Dica:
Calibrar os servos em posição 90° antes da montagem.

## 4. Programação (Arduino)
Código básico para movimentação:


```python
# Código exemplo (referência)
codigo = '''
#include <Servo.h>
Servo servo1;

void setup() {
  servo1.attach(2);
}

void loop() {
  servo1.write(90);
  delay(1000);
  servo1.write(45);
  delay(1000);
}
'''
print(codigo)
```

## 5. Controle de Movimento
O Otto utiliza padrões de movimento combinando ângulos dos servos:
- Andar
- Girar
- Dançar

Esses movimentos são sequências temporizadas.

## 6. Sensor Ultrassônico
Permite detectar obstáculos:

Distância = (tempo * velocidade do som) / 2

```python
# Simulação de leitura de distância
tempo = 0.01  # segundos
velocidade_som = 343  # m/s
distancia = (tempo * velocidade_som) / 2
distancia
```
