# Otto DIY — Modelo Cinemático Fiel

Notebook MATLAB/Octave inspirado no robô Otto DIY com:

- 4 servos reais
- Quadris com rotação em Z
- Pés com rotação em Y
- Movimento vertical do corpo
- Escala mais próxima do Otto real


```python
%% OTTO DIY - MODELO CINEMÁTICO FIEL (4 DOF)
% Quadris: rotação em Z
% Pés: rotação em Y
% Modelo inspirado na geometria real do Otto DIY
% https://ottodiy.com

clear;
close all;
clc;

%% =========================
% ESCALA REALISTA (mm)
%% =========================

bodyW = 80;
bodyH = 55;

hipSpacing = 40;

legL = 45;

footL = 55;
footH = 12;

baseY = 70;

%% =========================
% PARÂMETROS DA MARCHA
%% =========================

Ahip  = deg2rad(18);
Afoot = deg2rad(22);

w = 2*pi*0.7;

%% =========================
% FIGURA
%% =========================

figure('Color','w');

axis equal;
grid on;
hold on;

xlim([-120 120]);
ylim([0 180]);

xlabel('X');
ylabel('Y');

title('OTTO DIY - Modelo Cinemático Fiel');

%% =========================
% ANIMAÇÃO
%% =========================

for t = linspace(0,12,500)

    cla;

    %% ÂNGULOS

    hipL  =  Ahip*sin(w*t);
    hipR  = -Ahip*sin(w*t);

    footL_ang =  Afoot*sin(w*t + pi/2);
    footR_ang = -Afoot*sin(w*t + pi/2);

    %% ELEVAÇÃO DO CORPO

    bodyLift = 8*sin(w*t + pi/2);

    bodyY = baseY + legL + bodyLift;

    %% CORPO

    bodyX = 0;

    rectangle( ...
        'Position',...
        [bodyX-bodyW/2, bodyY, bodyW, bodyH],...
        'Curvature',0.2,...
        'FaceColor',[0.95 0.95 0.95],...
        'LineWidth',2);

    %% QUADRIS

    hipLeft  = [-hipSpacing/2, bodyY];
    hipRight = [ hipSpacing/2, bodyY];

    %% PERNAS

    legVecL = rot2d([0; -legL], hipL);
    legVecR = rot2d([0; -legL], hipR);

    ankleL = hipLeft  + legVecL';
    ankleR = hipRight + legVecR';

    plot([hipLeft(1) ankleL(1)], ...
         [hipLeft(2) ankleL(2)], ...
         'k','LineWidth',8);

    plot([hipRight(1) ankleR(1)], ...
         [hipRight(2) ankleR(2)], ...
         'k','LineWidth',8);

    %% PÉS

    footVecL = rot2d([footL/2; 0], footL_ang);
    footVecR = rot2d([footL/2; 0], footR_ang);

    footStartL = ankleL - footVecL';
    footEndL   = ankleL + footVecL';

    footStartR = ankleR - footVecR';
    footEndR   = ankleR + footVecR';

    plot([footStartL(1) footEndL(1)], ...
         [footStartL(2) footEndL(2)], ...
         'b','LineWidth',10);

    plot([footStartR(1) footEndR(1)], ...
         [footStartR(2) footEndR(2)], ...
         'b','LineWidth',10);

    %% JUNTAS

    plot(hipLeft(1), hipLeft(2), ...
         'ro','MarkerSize',10,'MarkerFaceColor','r');

    plot(hipRight(1), hipRight(2), ...
         'ro','MarkerSize',10,'MarkerFaceColor','r');

    plot(ankleL(1), ankleL(2), ...
         'go','MarkerSize',10,'MarkerFaceColor','g');

    plot(ankleR(1), ankleR(2), ...
         'go','MarkerSize',10,'MarkerFaceColor','g');

    %% CABEÇA/OLHOS

    eyeY = bodyY + bodyH*0.65;

    plot(-15, eyeY, 'ko', ...
         'MarkerSize',14,'MarkerFaceColor','k');

    plot(15, eyeY, 'ko', ...
         'MarkerSize',14,'MarkerFaceColor','k');

    %% TEXTO

    text(-45,165,...
        sprintf('t = %.2f s',t),...
        'FontSize',12);

    drawnow;

end

%% =========================
% FUNÇÃO AUXILIAR
%% =========================

function v2 = rot2d(v,ang)

R = [cos(ang) -sin(ang);
     sin(ang)  cos(ang)];

v2 = R*v;

end
```
